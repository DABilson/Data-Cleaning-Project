```python
#dropping the column 'date_first_booking'
airbnb_new = airbnb_info.drop('date_first_booking', axis = 1)
airbnb_new
```

The column 'date first booking' was dropped from the data due to insufficient data available as well as its irrelevance in the analysis it was to be used for.

```python
#changing Gender names to lower case
airbnb_new['gender'] = airbnb_new['gender'].str.lower()
airbnb_new
```

The name "Gender" was changed to a lower case to remain uniformity in the naming of the columns in the data set.

```python
#using strip method to remove '-' from 'unknown' under gender
airbnb_new['gender'] = airbnb_new['gender'].str.strip('-')
airbnb_new
```

The column gender has "-" attached to its naming and this could give error feedbacks during the analysis. Due to this, it was necessary to strip "-" from its name. 

```python
#replacing NaN values ('age') with median age
airbnb_new['age']=airbnb_new['age'].fillna(airbnb_new['age'].median())
airbnb_new
```

The 'Nan' values in the age column of the data set were filled with the median age of the entire ages in the column. This was necessary because the analysis could not be done with those missing values.

```python
#converting age from float to integer
airbnb_new['age']=airbnb_new['age'].astype(int)
airbnb_new
```

The age column data type was changed from a float to an integer. This is because age is always a whole number and not a decimal.

```python
#removing 'Other/' in 'Other/Unknown' under 'first device type'
airbnb_new['first_device_type'] = airbnb_new['first_device_type'].str.replace('Other/','')
airbnb_new
```

Under the "first_device_type", there are slashes preceeding the Unknown. This could not be worked with in the analysis therefore it was important to remove "Other/" from the data.

```python
#checking the unique values in 'first device type' 
airbnb_new['first_device_type'].unique()
```

This was done to check all the unique values in the column 'First device type'.
